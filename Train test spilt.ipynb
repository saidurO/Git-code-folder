```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split    
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
#feature scalling
df2 = df[['age', 'fare', 'parch', 'survived']]
df2.head()
```

```python
df3=df2.fillna(df2.mean())
df3.head()
```

```python
#x,y declaration
X=df3.drop('survived', axis=1)
Y=df3['survived']
X.head()
Y.head()
print('shape',X.shape)
print("shape of y",Y.shape)


```

```python
#Now we will do feature scalling using minmaxscaller and standardscaller
from sklearn.preprocessing import MinMaxScaler, StandardScaler
mms=MinMaxScaler()
X_mms=mms.fit_transform(X)
X_mms=pd.DataFrame(X_mms, columns=X.columns)
X_mms.head()
```

```python
#Train test spilt
X_train, X_test, Y_train, Y_test = train_test_split(X_mms, Y, test_size=0.2, random_state=42)
print("X_train shape=", X_train.shape)
print("X_test shape=", X_test.shape)
print("Y_train shape=", Y_train.shape)
print("Y_test shape=", Y_test.shape)
```
